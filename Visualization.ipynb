# Visualization with SEABORN and PLT

**FOCUS**: The following experiment will focus on three teams, three great teams: Real Madrid, Barcelona and Athletico Madrid. 

**GOAL**: This is purely for experimentation of the libraries **plt** and **seaborn**. There is no specific goal and the features were selected with no particular reason.


## Part 1: Import

In this experiment, the fifa database is used.

The following code, contains all the esential libraries to run the following experiment

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np

df = pd.read_csv("TRAIN.csv")
df.columns
```

## Part 2: Feature selection

Three teams were selected from the bigger database, each has 22 players.

```python
#Create new dataframes
dfReal = df[df.Club=="Real Madrid"]
dfBarc = df[df.Club=="FC Barcelona"]
dfAM = df[df.Club=="Atlético Madrid"]

#Merge them into a new dataframe, containging only the relevant information for the three teams.
DfThree = dfReal.append([dfBarc,dfAM])
```

## Part 3: Distribution

For the distributions, we check the overall score and how this is distributed for each of the three teams, for all of their players.

```python
#A nice, big space is created for the distributions
plt.figure(figsize=(16,6))

#The plots are placed
sns.distplot(a=dfReal['Overall'], label="Real Madrid", kde=False)
sns.distplot(a=dfBarc['Overall'], label="FC Barcelona", kde=False)
sns.distplot(a=dfAM['Overall'], label="Atlético Madrid", kde=False)

#Some asthetics
plt.title("Distributions for the three teams",color="green",fontsize=40)
plt.xlabel("Overall",fontsize=20,color="green")
plt.ylabel("Distributions",fontsize=20,color="green")
plt.legend()
```

## Part 4: Linear Relationship

The following script was created to show the correlation between two obviously, positive correlated features: Penalty and Finishing.

```python

#Create a nice big plane to plot
plt.figure(figsize=(16,6))
sns.regplot(x=DfThree['Finishing'], y=DfThree['Penalties'])
sns.set_style("darkgrid",{ 'figure.facecolor': 'black',  'xtick.color': 'Orange', 'ytick.color': 'Orange', "axes.labelcolor" : "Orange", 'axes.grid': True, 'grid.linestyle': '-',  'axes.edgecolor': 'Orange', 'grid.color': 'black',})

#Asthetics
plt.title("Corelation of two Features",color="blue",fontsize=40)
plt.xlabel("Finishing",fontsize=30,color="blue")
plt.ylabel("Penalties",fontsize=30,color="blue")
```

## Part 5: Scatterplot for best kickers

```python

#At this point, I am just enjoying it, playing with the colors, including additional parameter to check and boom!
plt.figure(figsize=(16,6))

sns.set_style("darkgrid",{ 'figure.facecolor': 'black',  'xtick.color': 'Orange', 'ytick.color': 'Orange', "axes.labelcolor" : "Orange", 'axes.grid': True, 'grid.linestyle': '-',  'axes.edgecolor': 'Orange', 'grid.color': 'black',})
sns.set_context("talk")
sns.scatterplot(x=DfThree['Finishing'], y=DfThree['Penalties'], hue=DfThree['Club'])

plt.title("Best Kickers",color="Blue",fontsize=40)
plt.xlabel("Finishing",fontsize=30,color="Blue")
plt.ylabel("Penalties",fontsize=30,color="Blue")
plt.legend(loc=4,fontsize = "medium")
```

## Part 6: Nationality per Club

```python

#And the game just got more exciting with segmenting the results for the three teams. Very cool stuff here
f, ax = plt.subplots(figsize=(15, 10))
sns.set_style("darkgrid")
sns.countplot(y="Nationality",hue="Club", data=DfThree, palette="Set2");
sns.set_style("darkgrid",{ 'figure.facecolor': 'black',  
                          'xtick.color': 'Orange', 
                          'ytick.color': 'Orange', 
                          'axes.grid': True, 
                          'grid.linestyle': '-',  
                          'axes.edgecolor': 'orange', 
                          'grid.color': 'black',})

plt.xlim(0,15)
plt.xticks(np.arange(0,15, 1))

plt.xlabel("Total Number of Footballers",fontsize=30,color="blue")
plt.ylabel("Nationality",fontsize=30,color="blue")
plt.title("Players' Nationality",color="blue",fontsize=40)
plt.legend(loc=4,fontsize = "x-large")
```
